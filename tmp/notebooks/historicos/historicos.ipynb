```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from dateutil.parser import parse
```

```python
s8lp = pd.read_csv("../../data/s8lp_historico.csv",index_col=0,parse_dates=True)
s8lp_d = s8lp.resample("D").min()
```

```python
sen0220 = pd.read_csv("../../data/sen0220_historico.csv",index_col=0,parse_dates=True)
sen0220_d = sen0220.resample("D").min()
```

```python
fig, ax = plt.subplots(figsize=(12,4))

# fecha1 = fecha - pd.Timedelta("15min")
# fecha2 = fecha1 + pd.Timedelta("25min")

ax.scatter(s8lp_d.index,s8lp_d,label='s8lp')
ax.scatter(sen0220_d.index,sen0220_d,label='sen0220')


# ax.set_xlim(fecha1,fecha2)
ax.set_ylim(300,850)
ax.grid()
ax.legend()
```
