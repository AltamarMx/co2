```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
file = '../../data/Disp_nuevos_prueba800.csv'
comerciales = pd.read_csv(file)
comerciales.columns = ['fecha','kCO2','DM1308B']
comerciales.fecha = '2021-07-30 ' + comerciales.fecha
comerciales.fecha = pd.to_datetime(comerciales.fecha,yearfirst=True)
comerciales.set_index('fecha',inplace=True)
sensores = comerciales.columns
```

```python
def importa_fluke(archivo):
    tmp = pd.read_csv(archivo,usecols=[7,11],names=["fluke","tiempo"],skiprows=1)
    tmp.tiempo = pd.to_datetime(tmp.tiempo,dayfirst=True)
    tmp.set_index("tiempo",inplace=True)
    tmp.fluke = tmp.fluke.str.replace(" ppm","")
    tmp.fluke = tmp.fluke.astype("float64")
    return tmp
```

```python
file = '../../data/fluje_30-07-2021_prueba800.csv'
fluke = importa_fluke(file)
# fluke
```

```python
fig, ax = plt.subplots(figsize=(20,8))


fecha1 = parse("2021-07-30 14:50") 
fecha2 = fecha1 + pd.Timedelta("60Min")
# fecha2 =  parse("2021-07-06 16:00") 
ax.set_xlim(fecha1,fecha2)

ax.set_ylim(400,800)
ax.plot(fluke,"k-o",label='fluke')
for sensor in sensores:
    ax.scatter(comerciales[sensor].index,comerciales[sensor],label=sensor,marker="o")

ax.grid()
ax.legend()
```
