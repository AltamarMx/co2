```python
tb = pd.read_csv("../data/sensores_26-07-2021_800ppm.csv",index_col=0,parse_dates=True)
# del tb["t3022"]
sensores = tb.columns
```

```python
inte = "1Min"
int2 = "10Min"
analiza_sensor(fluke,fluke,sensor="fluke",grafica=True,int_promedio=int2,intervalo=inte,imprime=False)
```

```python
for sensor in sensores:
    analiza_sensor(tb,fluke,sensor=sensor,grafica=True,intervalo='2Min',imprime=True)
```

```python
scd30    = analiza_sensor(tb,fluke,sensor="scd30",   grafica=False,int_promedio=int2,intervalo=inte)
sen0220  = analiza_sensor(tb,fluke,sensor="sen0220", grafica=False,int_promedio=int2,intervalo=inte)
s8lp     = analiza_sensor(tb,fluke,sensor="s8lp",    grafica=False,int_promedio=int2,intervalo=inte)
cjmcu811 = analiza_sensor(tb,fluke,sensor="cjmcu811",grafica=False,int_promedio=int2,intervalo=inte)
t3022    = analiza_sensor(tb,fluke,sensor="t3022",   grafica=False,int_promedio=int2,intervalo=inte)
resultados = pd.concat([scd30,sen0220,s8lp,cjmcu811,t3022],
                             axis=0)
resultados.to_csv("../data/resultado_800ppm.csv")

resultados
```
