```python
archivo = "../data/resultado_520ppm.csv"
r520ppm = pd.read_csv(archivo,index_col=[0,1])

archivo = "../data/resultado_625ppm.csv"
r625ppm = pd.read_csv(archivo,index_col=[0,1])

archivo = "../data/resultado_750ppm.csv"
r750ppm = pd.read_csv(archivo,index_col=[0,1])

archivo = "../data/resultado_800ppm.csv"
r800ppm = pd.read_csv(archivo,index_col=[0,1])

archivo = "../data/resultado_893ppm.csv"
r893ppm = pd.read_csv(archivo,index_col=[0,1])
```

```python
lista = [
#     r520ppm,
    r625ppm,
    r750ppm,
    r800ppm,
    r893ppm]
resultados  = pd.concat(lista,axis=0).sort_index(axis=0)
resultados
```

```python
promedio = resultados[["Error","TR+"]].groupby(by=['sensor']).agg(['mean','std'])
promedio.sort_values(by=("Error","mean"))
```
