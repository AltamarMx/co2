```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
from sensores import *
```

```python
archivo = "../data/fluke_26-07-2021_520ppm.csv"
fluke = pd.read_csv(archivo,index_col=0,parse_dates=True)
fluke.head(3)
```

```python
tb = pd.read_csv("../data/sensores_26-07-2021_520ppm.csv",index_col=0,parse_dates=True)
del tb["t3022"]
sensores = tb.columns
```

```python
inte = "600S"
int2 = "5Min"
analiza_sensor(fluke,fluke,sensor="fluke",grafica=True,int_promedio=int2,intervalo=inte,imprime=True)
```

```python
for sensor in sensores:
    analiza_sensor(tb,fluke,sensor=sensor,grafica=True,intervalo=inte,imprime=True)
```

```python

scd30    = analiza_sensor(tb,fluke,sensor="scd30",   grafica=False,int_promedio=int2,intervalo=inte)
sen0220  = analiza_sensor(tb,fluke,sensor="sen0220", grafica=False,int_promedio=int2,intervalo=inte)
s8lp     = analiza_sensor(tb,fluke,sensor="s8lp",    grafica=False,int_promedio=int2,intervalo=inte)
cjmcu811 = analiza_sensor(tb,fluke,sensor="cjmcu811",grafica=False,int_promedio=int2,intervalo=inte)
resultados_520ppm = pd.concat([scd30,sen0220,s8lp,cjmcu811],
                             axis=1,
                              keys=['scd30','sen0220','s8lp','cjmcu811'])
resultados_520ppm
```
