```python
datos = pd.read_csv("../data/pilot_study.csv",index_col=0,parse_dates=True)
```

```python
sensores = ['s8lp','scd30','sen0220']
fig, ax = plt.subplots(constrained_layout=True,figsize=(6,2.5))

fecha1 = parse("2021-09-07")
fecha2 = fecha1 + pd.Timedelta("4D")
ax.set_xlim(fecha1,fecha2)

ax.set_ylim(400,700)
ax.set_yticks(range(400,701,100))
ax.set_ylabel("CO2 [ppm]")
ax.set_xlabel("Time [MM:DD HH]")


ax.grid()

color = {"s8lp":"b-","sen0220":"g-","scd30":'r-'}
trans = {"s8lp":1 ,"sen0220":1 ,"scd30":1}
for sensor in sensores:
    if sensor=='flu'
    ax.plot(datos[sensor].index,datos[sensor],
            color[sensor],
            label=sensor,
            alpha=trans[sensor],markevery=1)

ax.legend()
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))


myFmt = mdates.DateFormatter('%m-%d %H:%M')
ax.xaxis.set_major_formatter(myFmt)
fig.savefig("../img/pilot_case.pdf",dpi=300)
```
