```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
import numpy as np
```

```python
tmp1150 = pd.read_csv('../data/calibracion_1150ppm.csv',index_col=0,parse_dates=True)
tmp650 = pd.read_csv('../data/calibracion_650ppm.csv',index_col=0,parse_dates=True)
tmp850 = pd.read_csv('../data/calibracion_850ppm.csv',index_col=0,parse_dates=True)
tmp450 = pd.read_csv('../data/calibracion_450ppm.csv',index_col=0,parse_dates=True)

calibracion = pd.concat([tmp1150,tmp650,tmp850,tmp450],axis=1)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

X = np.linspace(300,1200,2)
sensores = ['s8lp','scd30','sen0220']
colores  = ['C1','C2','C3']
for sensor,color in zip(sensores,colores):
#     ax.scatter(calibracion['fluke'],calibracion[sensor],s=0.02,c=color)
    ax.scatter(calibracion['fluke'].mean(),calibracion[sensor].mean()
               ,label=sensor,c=color)

ax.plot(X,X,'k-')
ax.plot(X,X+50,'g--')
ax.plot(X,X-50,'k--')




ax.set_yticks(np.arange(300,1250,100))
ax.set_xticks(np.arange(300,1250,100))
ax.set_xlabel("Fluke [ppm]")
ax.set_ylabel("Sensor [ppm]")
ax.legend()
ax.set_xlim(400,1200)
ax.set_ylim(400,1200)
ax.grid()
```
