```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
referencia = pd.read_csv("../data/referencia.csv",index_col=0,parse_dates=True)
sensores   = pd.read_csv("../data/sensores.csv",  index_col=0,parse_dates=True)
```

```python
datos = pd.concat([referencia,sensores],axis=1)
sensores = datos.columns
```

```python
fig, ax = plt.subplots(figsize=(10,6))

# fecha1 = parse("2021-06-21") 
# fecha2 = fecha1 + pd.Timedelta("2D")
# ax.set_xlim(fecha1,fecha2)

ax.set_ylim(300,1500)
ax.set_ylabel("CO2 [ppm]")
ax.set_xlabel("tiempo [mm-dd H]")
ax.set_title("Mediciones CO2")

for sensor in sensores:
    ax.scatter(datos[sensor].index,datos[sensor],label=sensor,alpha=.1)

ax.legend()
ax.grid()
```
