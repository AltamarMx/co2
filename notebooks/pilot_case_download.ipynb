```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dateutil.parser import parse
from tb import *
```

```python
tb = TB(config_file='config_gbv.ini',device_name='CO2')
```

```python
sensores = ["s8lp","sen0220","scd30"]
fecha1 = parse("2021-09-5")
fecha2 = fecha1 + pd.Timedelta("7D")
datos = pd.concat([tb.get_df(key=sensor,start_datetime=fecha1,end_datetime=fecha2) for sensor in sensores],axis=1)
datos.index = datos.index + pd.Timedelta("-5H")
# datos.tail()
```

```python
fig, ax = plt.subplots(figsize=(12,3))

# fecha1 = parse("2021-09-08")
# fecha2 = fecha1 + pd.Timedelta("48H")
ax.set_xlim(fecha1,fecha2)

ax.set_ylim(300,1000)
ax.set_ylabel("CO2 [ppm]")
ax.set_xlabel("tiempo")
ax.set_title("Sensores")
ax.grid()

color = {"s8lp":"b.","sen0220":"k.","scd30":'y.'}
trans = {"s8lp":0.3 ,"sen0220":0.01 ,"scd30":0.3}
for sensor in sensores:
    ax.plot(datos[sensor].index,datos[sensor],color[sensor],label=sensor,alpha=trans[sensor])

ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(12,3))

fecha1 = parse("2021-08-09 12:00")
fecha2 = fecha1 + pd.Timedelta("4H")
ax.set_xlim(fecha1,fecha2)

ax.set_ylim(300,1000)
ax.set_ylabel("CO2 [ppm]")
ax.set_xlabel("tiempo")
ax.set_title("Sensores")
ax.grid()

for sensor in sensores:
    ax.scatter(datos[sensor].index,datos[sensor],label=sensor)

ax.legend()
```
